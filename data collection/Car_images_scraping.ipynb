```python
import instaloader
import os

def scrape_instagram_images(profile):
    # Create an instance of Instaloader
    loader = instaloader.Instaloader()

    try:
        # Download profile's metadata
        loader.download_profile(profile, profile_pic=False)

        # Get the saved profile instance
        profile = instaloader.Profile.from_username(loader.context, profile)

        # Iterate over the profile's posts and download images
        for post in profile.get_posts():
            # Check if the post is an image
            if post.typename == 'GraphImage':
                loader.download_post(post, target=profile.username)
            else:
                # Skip videos and sidecar posts
                continue


        print("Image scraping completed!")
    except instaloader.exceptions.ProfileNotExistsException:
        print("Profile does not exist!")
    except Exception as e:
        print("An error occurred:", str(e))


```

```python
def remove_non_image_files(directory):
    for filename in os.listdir(directory):
        if not filename.endswith(".jpg") and not filename.endswith(".jpeg") and not filename.endswith(".png"):
            os.remove(os.path.join(directory, filename))
    print("Non-image files removed!")
```

```python
# Usage example
scrape_instagram_images("egyptprivateplate")
```

```python
remove_non_image_files('./egyptprivateplate/')
```
